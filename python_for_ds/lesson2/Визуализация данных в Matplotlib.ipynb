Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

```python
a = np.arange(0, 12)
print(f'a = {a}')
A = a.reshape(4, 3)
print(f'A = {A}')
At = A.transpose()
print(f'At = {At}')
B = A @ At
print(f'B = {B}')
print(f'B.size = {B.shape}')
det_B = np.linalg.det(B)
if det_B == 0:
    print(f'inverse matrix not exist')
else:
    print(f'inverse matrix = {np.linalg.inv(B)}')
```
